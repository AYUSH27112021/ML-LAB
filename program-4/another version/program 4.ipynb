```python
import pandas as pd
df=pd.read_csv(r"D:\github\ML\ML-LAB\program-4\New folder\data.csv")
df.head()
```

```python
df.drop('id',axis=1,inplace=True)#to drop id as its useless
df.head()
```

```python
#now we need to transform the dignosis
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df['diagnosis'].unique()
```

```python
df.shape
```

```python
feature_names=[i for i in df.columns]
feature_names
```

```python
X=df.iloc[:,:30]
Y=df.iloc[:,30:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=123)
```

```python
from sklearn.datasets import load_breast_cancer
from id3 import Id3Estimator
from id3 import export_graphviz
estimator = Id3Estimator()
estimator = estimator.fit(X_train, y_train)
y_pred=estimator.predict(X_test)
tree = export_graphviz(estimator.tree_, 'tree.dot', feature_names)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,target_names=['D','M']))
```
